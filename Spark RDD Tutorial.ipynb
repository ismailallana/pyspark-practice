```python
from pyspark import SparkConf, SparkContext

spark_conf = SparkConf()\
        .setAppName("RDD Tutorial")
sc=SparkContext.getOrCreate(spark_conf) 
```

```python
myRDD = sc.parallelize([('Ross',19),('Joey',18),('Rachel',16),('Phoebe',18),('Chandler',17),('Monica',20)])
```

```python
myRDD.collect()
```

```python
myRDD.take(4)
```

```python
def func(lines):
    lines = lines.lower()
    lines = lines.split()
    return lines
lower = New_RDD.map(func)
```

```python
lower.collect()
```

```python
rdd = New_RDD.flatMap(func)
```

```python
rdd.collect()
```

```python
a = sc.parallelize([('a',2),('b',3)])
```

```python
b = sc.parallelize([('a',3),('b',5),('c',2)]) 
```

```python
c = a.join(b)
```

```python
c.collect()
```

```python
c.reduceByKey(lambda x,y:x+y)
```

```python
c.collect()
```

```python
union = a.union(b)
```

```python
union.collect()
```

```python
union.reduceByKey(lambda x,y:x+y).collect()
```
